```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
fig, ax = plt.subplots(2,3,tight_layout=True)

x =np.arange(4)
y =[1,2,-5,2]
ax[0,0].plot(x,y)

x = np.linspace(0, 10, 1000)
y = np.sin(x)
ax[0,1].plot(x,y)

x=np.random.normal(5,2,1500)
ax[0,2].hist(x)

np.random.seed(5)
x = np.random.normal(20,5,50)
y = np.random.normal(15,7,len(x))
ax[1,0].scatter(x,y)

x = [np.random.randint(30, 50) for i in range(3)]
label =["","",""]
ax[1,1].pie(x, labels=label)

x = [np.random.randint(30, 1000) for i in range(100)]
ax[1,2].boxplot(x)

plt.show()
```
